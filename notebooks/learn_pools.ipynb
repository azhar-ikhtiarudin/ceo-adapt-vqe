```python
import numpy as np
from qiskit import QuantumCircuit

def qe_circuit(source_orbs, target_orbs, theta, n, big_endian=False):
    if len(source_orbs) == 2:
        return double_qe_circuit(source_orbs, target_orbs, theta, n, big_endian)
    else:
        return single_qe_circuit(source_orbs, target_orbs, theta, n, big_endian)
    

def single_qe_circuit(source_orb, target_orb, theta, n, big_endian=False):
    a = source_orb[0]
    b = target_orb[0]

    if big_endian:
        a = n - a - 1
        b = n - b - 1
    
    qc = QuantumCircuit(n)

    qc.rz(np.pi / 2, a)
    qc.rx(np.pi / 2, a)
    qc.rx(np.pi / 2, b)
    qc.cx(a, b)

    qc.rx(theta, a)
    qc.rz(theta, b)
    qc.cx(a, b)

    qc.rx(-np.pi / 2, b)
    qc.rx(-np.pi / 2, a)
    qc.rz(-np.pi / 2, a)

    return qc

def double_qe_circuit(source_orbs, target_orbs, theta, n, big_endian):

    a, b = source_orbs
    c, d = target_orbs

    if big_endian:
        # Qiskit's default is little endian - switch
        a = n - a - 1
        b = n - b - 1
        c = n - c - 1
        d = n - d - 1

    qc = QuantumCircuit(n)

    qc.cx(a, b)
    qc.cx(c, d)
    qc.x(b)
    qc.x(d)
    qc.cx(a, c)
    qc.ry(2 * theta / 8, a)

    qc.h(b)
    qc.cx(a, b)
    qc.h(d)
    qc.ry(-2 * theta / 8, a)

    qc.cx(a, d)
    qc.ry(2 * theta / 8, a)

    qc.cx(a, b)
    qc.h(c)
    qc.ry(-2 * theta / 8, a)

    qc.cx(a, c)
    qc.ry(2 * theta / 8, a)

    qc.cx(a, b)
    qc.ry(-2 * theta / 8, a)

    qc.cx(a, d)
    qc.ry(2 * theta / 8, a)

    qc.cx(a, b)
    qc.ry(-2 * theta / 8, a)

    qc.h(d)
    qc.h(b)
    qc.rz(+np.pi / 2, c)
    qc.cx(a, c)

    qc.rz(-np.pi / 2, a)
    qc.rz(+np.pi / 2, c)
    qc.ry(+np.pi / 2, c)

    qc.x(b)
    qc.x(d)
    qc.cx(a, b)
    qc.cx(c, d)

    return qc
```

```python
# Obtain the QE Circuit

from qiskit import QuantumCircuit

circuit = QuantumCircuit(pool.n)

indices = [2, 1]
coefficients = [0.1235, 0.44]

for i, (index, coefficient) in enumerate(zip(indices, coefficients)):
    operator = pool.operators[index]
    print(operator)
    source_orbs = operator.source_orbs
    target_orbs = operator.target_orbs
    print("\n",source_orbs, target_orbs)

    qc = qe_circuit(source_orbs, target_orbs, coefficient, pool.n, big_endian=True)

    print("\n ======= QE Circuit ", i, "======= ")
    print(qc)
    circuit = circuit.compose(qc)
    circuit.barrier()

print("\nFinal QE Circuit:")
print(circuit)

```

```python
# Obtain the Paramterized QE Circuit

from qiskit import QuantumCircuit
from qiskit.circuit import ParameterVector

circuit = QuantumCircuit(pool.n)

indices = [2, 1]
coefficients = [0.1235, 0.44]
parameters = ParameterVector("theta", len(indices))
print(parameters)

for i, (index, coefficient) in enumerate(zip(indices, coefficients)):
    operator = pool.operators[index]
    print(operator)
    source_orbs = operator.source_orbs
    target_orbs = operator.target_orbs
    print("\n",source_orbs, target_orbs)

    qc = qe_circuit(source_orbs, target_orbs, parameters[i], pool.n, big_endian=True)

    print("\n ======= QE Circuit ", i, "======= ")
    print(qc)
    circuit = circuit.compose(qc)
    circuit.barrier()

print("\nFinal QE Circuit:")
print(circuit)

```
